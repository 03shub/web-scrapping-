```python
import pandas as pd
from selenium import webdriver
import selenium 
import time  
from selenium.common.exceptions import NoSuchElementException
```

```python
url = "https://www.naukri.com/"
```

```python
driver.get(url)
```

```python
job_search = driver.find_element_by_class_name("sugInp")
```

```python
job_search.send_keys("data analyst")
```

```python
location_search = driver.find_element_by_id("qsb-location-sugg")
```

```python
location_search.send_keys("banglore")
```

```python
btn = driver.find_element_by_class_name("btn")
```

```python
btn.click()
```

```python
urls = []
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    urls.append(i.get_attribute("href"))
    
```

```python
len(urls)
```

```python
urls
```

```python
job_title = []
job_discription = []
```

```python
for i in urls[:10]:
    driver.get(i)
    time.sleep(3)
    try:
        job=driver.find_element_by_xpath("//h1[@class='jd-header-title']")
        job_title.append(job.text.replace("/n"),"new line")
    except:
        job_title.append('-')
        
    try:
        job_desc = driver.find_element_by_xpath("//section[@class='job-desc']")
        job_discription.append(job_desc.text)
    except:
            job_discription.append('-')
```

```python
job_discription
```

```python
job_title
```

```python
job = pd.DataFrame({})
job["job_discription"]=job_discription
job["job_title"]=job_title
```

```python
job
```
